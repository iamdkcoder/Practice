# graph construction

1. Adjacency Matrix
an adjacency matrix is a 2d array of V x V vertices. Each row and column represent a vertex 


```python
graph = [[0,1] ,[1,2]]
v=4

matrix = [[0]*3 for i in range(3)]

# this is for undirected relationships
for rel in graph:
    u,v = rel 
    matrix[u][v]=1
    matrix[v][u]=1


print(matrix)
```

2. Adjacency list 

An adjacency list represents a vertex and each element in its linked list represents with other vertices that form an edge with the vertex



```python
graph = [[0,1] ,[1,2]]

graph_hm = {}


for rel in graph:
    u,v = rel
    uList  = graph_hm.get(u,[])
    uList.append(v)
    graph_hm[u]= uList

    vList = graph_hm.get(v,[])
    vList.append(u)

    graph_hm[v] = vList

print(graph_hm)
```
