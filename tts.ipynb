```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-hin")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-hin")

text = "द्वापर युग धर्म की रक्षा के लिए महाभारत का युद्ध लड़ा गया था। कुरुक्षेत्र में लड़े जाने के कारण इसे कुरुक्षेत्र का युद्ध भी कहा जाता है। इस युद्ध की सबसे विशेष बात यह थी कि इसमें भगवान श्री कृष्ण स्वयं अर्जुन के सारथी बने थे। यह भयावह युद्ध कुल 18 दिनों तक चला।"
text ='रतन टाटा ने फिर कभी शादी नहीं की। शायद ये उनकी अधूरी मोहब्बत थी, जो दिल में कहीं गहरी छुपी रह गई। लेकिन ये अधूरी कहानी भी उनके मजबूत और सच्चे इंसान होने का सबूत देती है।'
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```
